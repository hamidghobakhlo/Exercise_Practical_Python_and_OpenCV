## Imread

```python
import cv2
import matplotlib.pyplot as plt

# مسیر تصویر
image_path = "./images/bird.jpg"

# خواندن تصویر با OpenCV
image = cv2.imread(image_path)

# بررسی بارگذاری تصویر
if image is None:
    print("Error!")
else:
    # تبدیل BGR به RGB (چون OpenCV تصویر را در فرمت BGR می‌خواند، اما matplotlib نیاز به RGB دارد)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # نمایش تصویر با matplotlib
    plt.figure(figsize=(8, 6))
    plt.imshow(image_rgb)
    plt.title("Original Image")
    plt.axis("off")  # برای مخفی کردن محورها
    plt.show()
```
